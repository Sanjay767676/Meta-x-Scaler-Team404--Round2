```python
# Verify required submission artifacts exist
from pathlib import Path
required_outputs = [
    'outputs/reward_curve.png',
    'outputs/pass_rate.png',
    'outputs/loss_curve.png',
    'outputs/final_report.json',
]
required_logs = [
    'logs/summary.json',
    'logs/episodes.csv',
]
for file_path in required_outputs + required_logs:
    p = Path(file_path)
    print(f'{file_path}:', 'OK' if p.exists() else 'MISSING')
```

```python
# Display generated evidence charts
from pathlib import Path
import matplotlib.pyplot as plt
from PIL import Image

chart_paths = [
    Path('outputs/reward_curve.png'),
    Path('outputs/pass_rate.png'),
    Path('outputs/loss_curve.png'),
]

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for ax, path in zip(axes, chart_paths):
    if path.exists():
        ax.imshow(Image.open(path))
        ax.set_title(path.name)
        ax.axis('off')
    else:
        ax.set_title(f'{path.name} (missing)')
        ax.axis('off')
plt.tight_layout()
plt.show()
```
