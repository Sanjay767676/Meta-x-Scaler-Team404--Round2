```python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
import torch

base_model = "Qwen/Qwen2.5-Coder-1.5B-Instruct"

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4"
)

print("Loading tokenizer...")
tokenizer = AutoTokenizer.from_pretrained(base_model)

print("Loading base model...")
model = AutoModelForCausalLM.from_pretrained(
    base_model,
    device_map="auto",
    quantization_config=bnb_config
)

print("Qwen 1.5B model ready.")
```

```python
prompt = """
Write Python code to sort an integer array safely with negatives.
Return only code.
"""

inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

outputs = model.generate(
    **inputs,
    max_new_tokens=200,
    temperature=0.7,
    do_sample=True,
    pad_token_id=tokenizer.eos_token_id
)

response = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(response)
```

```python
prompt = """
You are a coding assistant.

Return ONLY valid Python code.

Task:
Write function:

def solve(arr):

that returns the sorted integer list.
Must support negatives, duplicates and empty arrays.
No explanation.
"""

inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

outputs = model.generate(
    **inputs,
    max_new_tokens=120,
    temperature=0.1,
    do_sample=False,
    pad_token_id=tokenizer.eos_token_id
)

print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```

```python
from unsloth import FastLanguageModel

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="Qwen/Qwen2.5-Coder-1.5B-Instruct",
    max_seq_length=2048,
    load_in_4bit=True,
)

model = FastLanguageModel.get_peft_model(
    model,
    r=16,
    target_modules=[
        "q_proj",
        "k_proj",
        "v_proj",
        "o_proj"
    ],
    lora_alpha=16,
    lora_dropout=0,
    bias="none",
    use_gradient_checkpointing="unsloth"
)

print("LoRA attached successfully")
```

```python
from datasets import Dataset

ds = Dataset.from_list([
{"text":"def solve(arr): return sorted(arr)"},
{"text":"def solve(arr): return sorted(arr)"},
{"text":"def solve(arr): return sorted(arr)"}
])
```

```python
from trl import SFTTrainer, SFTConfig

trainer = SFTTrainer(
    model=model,
    processing_class=tokenizer,
    train_dataset=ds,
    args=SFTConfig(
        output_dir="out",
        max_steps=10,
        per_device_train_batch_size=1,
        logging_steps=1,
        report_to="none",
        dataset_text_field="text"
    )
)

trainer.train()
```

```python
# FORGE-v4 REAL DATA ONLY Colab Validator
# Uses ONLY real outputs from:
# 1. Your Hugging Face Space endpoints
# 2. Your local FORGE repo benchmark logs
# No hardcoded/fake metrics

!pip -q install requests pandas matplotlib

import requests
import pandas as pd
import matplotlib.pyplot as plt
import subprocess
import json
import os
import re
from IPython.display import display

# ---------------------------------------------------
# CONFIG
# ---------------------------------------------------
SPACE_URL = "https://sanjay7676-team404-forge.hf.space"
EPISODES = 10

# ---------------------------------------------------
# PART 1 — REAL SPACE CALLS
# ---------------------------------------------------
print("🌐 Testing Live FORGE Space (real endpoints)\n")

def call_endpoint(method, url, payload=None):
    try:
        if method == "GET":
            r = requests.get(url, timeout=60)
        else:
            r = requests.post(url, json=payload, timeout=60)

        print(f"✅ {url}")
        print("Status:", r.status_code)
        print(r.text[:2000])
        print("-" * 60)
        return r
    except Exception as e:
        print(f"❌ {url}")
        print(e)
        print("-" * 60)
        return None

# reset
call_endpoint("POST", f"{SPACE_URL}/reset")

# state
call_endpoint("GET", f"{SPACE_URL}/state")

# step
step_payload = {
    "action": {
        "code": "def solve(arr): return sorted(arr)"
    }
}
call_endpoint("POST", f"{SPACE_URL}/step", step_payload)

# ---------------------------------------------------
# PART 2 — RUN REAL LOCAL BENCHMARK
# ---------------------------------------------------
print("\n🚀 Running real benchmark from your codebase...\n")

if os.path.exists("train_colab.py"):
    cmd = f"python train_colab.py --compare --episodes {EPISODES}"

    process = subprocess.Popen(
        cmd.split(),
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT,
        text=True
    )

    raw_logs = []

    for line in process.stdout:
        print(line, end="")
        raw_logs.append(line)

    process.wait()

else:
    print("❌ train_colab.py not found.")

# ---------------------------------------------------
# PART 3 — USE REAL CSV LOGS ONLY
# ---------------------------------------------------
csv_path = "logs/episodes.csv"

if os.path.exists(csv_path):
    print("\n📂 Found real logs:", csv_path)

    df = pd.read_csv(csv_path)
    display(df.head())

    # detect reward column
    reward_col = next((c for c in df.columns if "reward" in c.lower()), None)

    if reward_col:
        plt.figure(figsize=(10,5))
        plt.plot(df[reward_col], marker="o")
        plt.title("FORGE Real Reward Curve")
        plt.xlabel("Episode")
        plt.ylabel("Reward")
        plt.grid(True)
        plt.show()

    # detect pass column
    pass_col = next((c for c in df.columns if "pass" in c.lower()), None)

    if pass_col:
        plt.figure(figsize=(10,5))
        plt.plot(df[pass_col], marker="o")
        plt.title("FORGE Real Pass Rate")
        plt.xlabel("Episode")
        plt.ylabel("Pass Rate")
        plt.grid(True)
        plt.show()

else:
    print("\n⚠️ logs/episodes.csv not found.")
    print("No graph shown because fake data is not allowed.")

# ---------------------------------------------------
# PART 4 — REAL JSON REPORTS
# ---------------------------------------------------
for report in [
    "outputs/final_report.json",
    "outputs/summary.json",
    "outputs/TRAINING_REPORT.md"
]:
    if os.path.exists(report):
        print(f"\n📄 Found real file: {report}")

# ---------------------------------------------------
# FINAL
# ---------------------------------------------------
print("\n🏆 ONLY REAL DATA DISPLAYED")
print("✔ No fake numbers used")
print("✔ Graphs shown only if real logs exist")
print("✔ Space endpoints called live")
print("✔ Benchmark data taken from your actual repo")
```
